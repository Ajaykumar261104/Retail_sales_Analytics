```python
merged_df = pd.merge(orders_df, order_items_df, on='order_id')
```

```python
merged_df['TotalAmount'] = (
    merged_df['quantity'] *
    merged_df['list_price'] *
    (1 - merged_df['discount'])
)
```

```python
merged_df
```

```python
sales_df = pd.merge(
    orders_df,
    order_items_df,
    on="order_id",
    how="inner"
)
```

```python
sales_df["order_date"] = pd.to_datetime(
    sales_df["order_date"],
    dayfirst=True
)
```

```python
sales_df["Revenue"] = (
    sales_df["list_price"] *
    sales_df["quantity"] *
    (1 - sales_df["discount"])
)
```

```python
reference_date = sales_df["order_date"].max() + pd.Timedelta(days=1)
```

```python
rfm_data = sales_df.groupby("customer_id").agg(
    Recency=("order_date", lambda x: (reference_date - x.max()).days),
    Frequency=("order_id", "nunique"),
    Monetary=("Revenue", "sum")
).reset_index()
```

```python
rfm_data
```

```python
#Task4
```

```python
def customer_segment(monetary):
    if monetary > 50000:
        return "High Value"
    elif monetary >= 10000:
        return "Medium Value"
    else:
        return "Low Value"

rfm_data["Segment"] = rfm_data["Monetary"].apply(customer_segment)

print(rfm_data.head())    
```
